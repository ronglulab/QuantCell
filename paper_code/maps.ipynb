```python
from maps.cell_phenotyping import Trainer
from maps.cell_phenotyping import Predictor
import pandas as pd
import os
import time

base_dir = '/Projects/wboohar/PhenoCycler'
project_name = 'QuantCellPaper'

project_path = f'{base_dir}/{project_name}'

results_dir = f'{project_path}/MAPS/results/cell_phenotyping/'
os.makedirs(results_dir, exist_ok=True)

train_data_path = f'{project_path}/MAPS/data/cell_phenotyping/train_features.csv'
valid_data_path = f'{project_path}/MAPS/data/cell_phenotyping/valid_features.csv'

model = Trainer(results_dir=results_dir, num_features=26, num_classes=21, verbose=1, min_epochs=5, patience=25)
```

```python
start = time.time()

model.fit(train_data_path, valid_data_path)

pretrained_model_checkpoint_path = f'{project_path}/MAPS/results/cell_phenotyping/best_checkpoint.pt'
data_path = f'{project_path}/MAPS/data/cell_phenotyping/test_features.csv'

model = Predictor(model_checkpoint_path=pretrained_model_checkpoint_path, num_features=26, num_classes=21)
pred_labels, pred_probs = model.predict(data_path)

end = time.time()
time_elapsed = end - start

pd.Series(pred_labels).to_csv(f'{project_path}/MAPS/results/cell_phenotyping/pred_labels_features.csv', index=True)
pd.DataFrame(pred_probs).to_csv(f'{project_path}/MAPS/results/cell_phenotyping/pred_probs_features.csv', index=True)
pd.Series(time_elapsed).to_csv(f'{project_path}/MAPS/results/cell_phenotyping/time_elapsed_maps_features.csv', index=True)
```
